```python
import numpy as np
import timeit

print(np.sum(np.arange(4)))

print("Time taken by vectorized sum :",end= "")
%timeit np.sum(np.arrange(4))

total = 0
for item in range(0,4):
    
    total += item 
a = total
print("\n" + str(a))

print("Time taken by iterative sum :",end= "")
%timeit a
```
